```python
import pandas as pd 
import numpy as np 

path = "data/Honoraires_totaux_des_professionnels_de_sante_par_departement_en_2013.xls"
df = pd.read_excel(path,sheetname=[x for x in xrange(2,6)])
#quand  sheetname est une liste, la méthode retourne un dictionnaire de dataframe où chaque clé représente une feuille excel
df[2][df[2]['DEPASSEMENTS (euros)'] == 'nc']
```

```python
# nettoyer le dataframe 
df[2] = df[2].replace('nc', np.nan).dropna()
df[2]
```

```python
import re
depar_nums = []
for x in df[2]['DEPARTEMENT']:
    m= re.split('-',x)
    if len(m[0]) <4: 
        depar_nums.append(m[0])
#depar_nums
value = '01-blabla'
m= re.split('-',value)
m[1]
```

```python
subframe = df[2].loc [:, ['SPECIALISTES','DEPARTEMENT','EFFECTIFS','DEPASSEMENT MOYEN (euros)']]
subframe.head(5)
```

```python
subframe = subframe[subframe['EFFECTIFS'] <> 0]
subframe.head(5)
```

```python
#test = pd.DataFrame(subframe['DEPARTEMENT'].apply(lambda x : x.split('-')))
#test = pd.DataFrame(subframe['DEPARTEMENT'].str.split('-'), columns = ['first','secod'])
test = subframe['DEPARTEMENT'].apply(lambda x: pd.Series(re.split('-',x)))
testfinal = test[1] +' '+  test[2] +' '+ test [3]

#dataframe["period"] = dataframe["Year"].map(str) + dataframe["quarter"]
```

```python
subframe =  subframe[subframe['DEPARTEMENT'].str.contains('TOTAL*') == False]
subframe.head(5)
```

```python
subframe =  subframe[subframe['SPECIALISTES'].str.contains('TOTAL*') == False]
subframe.head(5)
```

```python
subframe[subframe['DEPARTEMENT']=='01- Ain']
effe_dep_df = pd.DataFrame(subframe.groupby(['DEPARTEMENT'], as_index=False)['EFFECTIFS'].sum())
effe_dep_df.head(5)
```

```python
effe_dep_tot_df = pd.merge(subframe, effe_dep_df, how='inner', on=['DEPARTEMENT'])
effe_dep_tot_df['DENSITY'] = effe_dep_tot_df['EFFECTIFS_x']/effe_dep_tot_df['EFFECTIFS_y']
effe_dep_tot_df.head(10)
```

```python
#plot le dépassement moyen en fonction de la densité 
%matplotlib inline 
import matplotlib.pyplot as plt
for key, group in effe_dep_tot_df.groupby(['DEPARTEMENT']):
    plt.bar(group["DEPASSEMENT MOYEN (euros)"].head(100), group["DENSITY"].head(100),10, alpha=0.4)
    plt.show()
```

```python
sum_all_dep = effe_dep_df['EFFECTIFS'].sum()
effe_dep_df['DENSITY'] = effe_dep_df['EFFECTIFS'] / sum_all_dep
effe_dep_df.head(5)
```

```python
subframe_2 = df[2].loc [:, ['DEPARTEMENT','DEPASSEMENTS (euros)','NOMBRE DE DEPASSEMENTS']]
sum_df = pd.DataFrame(subframe_2.groupby(['DEPARTEMENT'], as_index=False )['DEPASSEMENTS (euros)','NOMBRE DE DEPASSEMENTS'].sum())
sum_df['DEPASSEMENT MOYEN'] = sum_df['DEPASSEMENTS (euros)']/sum_df['NOMBRE DE DEPASSEMENTS']
sum_df.head(5)
```

```python
all_spec_df = pd.merge(sum_df,effe_dep_df , how='inner', on=['DEPARTEMENT'])
all_spec_df['DENSITY'] = all_spec_df['DENSITY']*1000
all_spec_df.head(5)
```

```python
index = np.arange(10)
bar_width = 0.35

opacity = 0.4

v1 = tuple(all_spec_df['DEPASSEMENT MOYEN'][:10].tolist())
v2 = tuple(all_spec_df['DENSITY'][:10].tolist())

rects1 = plt.bar(index, v1, bar_width,
                 alpha=opacity,
                 color='g',
                 label='DEPASSEMENT MOYEN')

rects2 = plt.bar(index + bar_width, v2, bar_width,
                 alpha=opacity,
                 color='y',
                 label='DENSITY')

plt.xlabel('DEPASSEMENT MOYEN')
plt.ylabel('DENSITY')
plt.title('DEPASSEMENT AND DENSITY / DEPARTEMENT')
plt.xticks(index + bar_width, tuple(all_spec_df['DEPARTEMENT'][:10].tolist()))
plt.axis([0,5,0,70])
plt.legend()
#plt.figure(num=None, figsize=(25, 15), dpi=80, facecolor='w', edgecolor='k')
plt.tight_layout()
plt.show()
```

```python
#Les données concernant la densité de la population par tranche d'age 
path = "data/data_pop.csv"
pop_df = pd.read_csv(path, sep=";")
#garder que les données de l'année 2013
pop_df = pop_df[pop_df['annee']==2013]
pop_df.head(5)
```

```python
sum_value_dep_sexe_df = pd.DataFrame(pop_df.groupby(['l_dep','sexe'], as_index=False)['value'].sum())
sum_value_dep_sexe_df.head(10)
```

```python
fi_df = pd.merge(pop_df,sum_value_dep_sexe_df, how='inner', on=['l_dep','sexe'])
fi_df['density'] = fi_df['value_x']/fi_df['value_y']
fi_df.head(10)
```
