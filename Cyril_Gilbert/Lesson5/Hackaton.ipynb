```python
import re
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
import unicodedata
import matplotlib.pylab as plt

def getContentFromUrl(url):
	# Execute a request to get the content from a web page
	request = requests.get(url)
	# Parse the document
	soup = BeautifulSoup(request.text, 'html.parser')
	return soup

def clean_list_with_number(liste):
    new_colonne = []
    for nom_colonne in liste:
        nom_inter = unicodedata.normalize('NFKD', nom_colonne).encode('ascii','ignore')
        nom_inter = nom_inter.replace("- ","")
        nom_inter = nom_inter[2:]
        new_colonne.append(nom_inter)
    return new_colonne

def clean_list(liste):
    new_colonne = []
    for nom_colonne in liste:
        nom_inter = unicodedata.normalize('NFKD', nom_colonne).encode('ascii','ignore')
        nom_inter = nom_inter.replace("- ","")
        new_colonne.append(nom_inter)
    return new_colonne
```

```python
#DATAFRAME DES DEPASSEMENTS D'HONORAIRES###

df = pd.read_excel('Honoraires_totaux_des_professionnels_de_sante_par_departement_en_2013.xls',sheetname=2)
df = df.replace('nc',np.nan)
df = df.dropna()
df = df[df['DEPARTEMENT'].str.contains('TOTAL')==False]
df = df[df['SPECIALISTES'].str.contains('TOTAL')==False]

df_spe_eff_depart = df[['SPECIALISTES','DEPARTEMENT','DEPASSEMENTS (euros)']]
#df_spe_eff_depart = df_spe_eff_depart.dropna()
df_spe_eff_depart = df_spe_eff_depart.pivot('DEPARTEMENT','SPECIALISTES','DEPASSEMENTS (euros)')
df_spe_eff_depart
new_colonne = clean_list_with_number(df_spe_eff_depart.columns)
df_spe_eff_depart.columns = new_colonne

new_index = clean_list_with_number(df_spe_eff_depart.index)
df_spe_eff_depart.index = new_index

df_spe_eff_depart = df_spe_eff_depart[sorted(df_spe_eff_depart.columns)]

df_spe_eff_depart.drop(df_spe_eff_depart.columns[0], axis=1, inplace=True)

df_spe_eff_depart.head()
```
