```python
import pandas as pd
```

```python
# source file: http://www.insee.fr/fr/themes/detail.asp?reg_id=99&ref_id=pop-sexe-age-quinquennal
file_pop = '/cal/homes/piochaud/workpython/hackaton_2015/data/pop-sexe-age-quinquennal6812_3.xls'
skiprows = range(12) + [13]
sheetname = 'COM_2012'
data_pop = pd.read_excel(file_pop, skiprows=skiprows, sheetname=sheetname)
```

```python
from mpl_toolkits.basemap import Basemap
import numpy
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(1, 1, figsize=(8,8))
m = Basemap(llcrnrlon=-5,llcrnrlat=40,urcrnrlon=20,urcrnrlat=56,
            resolution='i',projection='cass',lon_0=2.34,lat_0=48,
           ax=axes)
m.drawcoastlines()
m.drawcountries()
m.fillcontinents(color='lightgrey', lake_color='#AAAAFF')
m.drawparallels(numpy.arange(-40,61.,2.))
m.drawmeridians(numpy.arange(-20.,21.,2.))
m.drawmapboundary(fill_color='#BBBBFF')
```
