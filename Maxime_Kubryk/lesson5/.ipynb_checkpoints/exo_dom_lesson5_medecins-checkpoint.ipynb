```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
from matplotlib import cm
import seaborn as sns
from binning import *
from pandas.tools.plotting import radviz, scatter_matrix
import collections
%matplotlib inline
```

#### On commence par nettoyer les données

```python
# https://www.data.gouv.fr/fr/datasets/honoraires-des-professionnels-de-sante-liberaux/
path = "data/Honoraires_totaux_des_professionnels_de_sante_par_departement_en_2013.xls"

data = pd.read_excel(path, sheetname=[1, 2, 3, 4, 5, 6])

df = data[2]

# enlever les nan
df = df.replace('nc', np.nan).dropna()

# extraire les codes numeriques
df = df[df['DEPARTEMENT'].str.contains('- ')]
df = df[df['SPECIALISTES'].str.contains('- ')]

dep = pd.DataFrame([[x] + x.split('- ') for x in df['DEPARTEMENT']],
                       columns=['DEPARTEMENT', 'num_dep', 'name_dep'])

spec = pd.DataFrame([[x] + x.split('- ') for x in df['SPECIALISTES']],
                        columns=['SPECIALISTES', 'num_spec', 'name_spec'])

df = pd.concat([df, dep, spec], axis=1, join='inner')

df['num_dep'] = df['num_dep'].str.replace('^0', '').str.replace(
        'B', '.5').str.replace('A', '.25').astype('float')

df['num_spec'] = df['num_spec'].replace('^0', '').astype('float')
df['DEPASSEMENTS (euros)'] = df['DEPASSEMENTS (euros)'].astype('float')

# ne garder que les lignes avec EFFECTIF > 0
df = df[df['EFFECTIFS'] > 0]



print(df.head())

# des quantités qui pourront être utiles
#df['NOMBRE DE DEPASSEMENTS (/medecin)'] = df['NOMBRE DE DEPASSEMENTS'] / df['EFFECTIFS']
#df['DEPASSEMENTS (euros/medecin)'] = df['DEPASSEMENTS (euros)'] / df['EFFECTIFS']
```

#### scatter matrix

```python
dat = df[['num_spec', 'NOMBRE DE DEPASSEMENTS','DEPASSEMENT MOYEN (euros)', 'DEPASSEMENTS (euros)', 'EFFECTIFS']]

plt.style.use('ggplot')
scatter_matrix(dat, diagonal='kde',figsize=(15,13))
```

#### correlation entre les variables

```python
dat = df[['num_spec', 'NOMBRE DE DEPASSEMENTS',
              'DEPASSEMENT MOYEN (euros)', 'DEPASSEMENTS (euros)', 'EFFECTIFS']]
corr_mat = dat.corr()

plt.figure(figsize=(9,9))
sns.heatmap(corr_mat, square=True, vmax=0.8)
```

La matrice des corrélations est pas très intéressante... 

```python
plt.figure(figsize=(12,9))
radviz(dat, 'num_spec')
```

Dans ce graphe, chaque point est attiré par les 4 pôles (par exemple si un point a un 'effectifs' très grand et si les autres grandeurs sont faibles, ce point sera très proche du pôle 'effectifs).

Sur ce graphe, il semble que pour les specialités dont les effectifs sont généralement élevés (par département) les dépassements sont généralement faibles (par exemple la spécialité 75). Et inversement (par exemple la spécialité 79). 

Par ailleurs, le nombre de dépassements et le dépassement total étant liés approximativement linéairement, ces deux grandeurs ont peu d'influence dans ce graphe.

```python
# spécialité 75
dat =  df[df['num_spec']==75]
name = str(dat['num_spec'].unique()[0])+" "+dat['name_spec'].unique()[0]
ax = dat.plot(kind='scatter',x='EFFECTIFS',y='DEPASSEMENT MOYEN (euros)',
              figsize=(12,8),label=name)


# spécialité 79
dat = df[df['num_spec']==79]
name = str(dat['num_spec'].unique()[0])+" "+dat['name_spec'].unique()[0]
dat.plot(kind='scatter',x='EFFECTIFS',y='DEPASSEMENT MOYEN (euros)',
                            color='r',ax=ax,label=name)

# spécialité 78
dat = df[df['num_spec']==78]
name = str(dat['num_spec'].unique()[0])+" "+dat['name_spec'].unique()[0]
dat.plot(kind='scatter',x='EFFECTIFS',y='DEPASSEMENT MOYEN (euros)',
                            color='k',ax=ax,label=name)

# spécialité 43
dat = df[df['num_spec']==77]
name = str(dat['num_spec'].unique()[0])+" "+dat['name_spec'].unique()[0]
dat.plot(kind='scatter',x='EFFECTIFS',y='DEPASSEMENT MOYEN (euros)',
                            color='orange',ax=ax,label=name)

```

```python
groups = df.groupby(['name_spec'])

plt.figure(figsize=(12,9))
plt.title('par specialité')
plt.xlabel('effectifs moyen (moyenne sur les departements)')
plt.ylabel('depassement moyen (moyenne sur les departemt.)')
x = []
y = []
for name, group in groups:
    avg_eff = group['EFFECTIFS'].mean()
    avg_depassementmoyen = group['DEPASSEMENT MOYEN (euros)'].mean()
    x.append(avg_eff)
    y.append(avg_depassementmoyen)
    plt.plot(avg_eff, avg_depassementmoyen,'o',label=name)
    
bins = linspace(min(x),max(x),9)
count = binning_1D(bins,x,5,2,1) 
binning = binning_1D(bins,x,5,2,y)

smooth = binning/count
plt.plot(bins,smooth,'r',lw=2)


c = np.column_stack((np.asarray(x),np.asarray(y)))
data = pd.DataFrame(c, columns=['Effectifs en moyens','Depassements moyens'])

sns.jointplot('Effectifs en moyens','Depassements moyens', data=data, kind="reg", color="r", size=10)

    
#plt.legend(loc='upper right',ncol=3)
    
```

On voit qu'en moyenne sur tous les départements, plus une spécialité est rare plus les dépassements pratiqués peuvent être elevés.

```python
groups = df.groupby(['name_dep'])

plt.figure(figsize=(12,9))
plt.title('par departement')
plt.xlabel('effectifs moyen')
plt.ylabel('depassement moyen (sur toutes les disciplines)')
x = []
y = []
for name, group in groups:
    avg_eff = group['EFFECTIFS'].mean()
    avg_depassementmoyen = group['DEPASSEMENT MOYEN (euros)'].mean()
    x.append(avg_eff)
    y.append(avg_depassementmoyen)
    plt.plot(avg_eff, avg_depassementmoyen,'o',label=name)
    
bins = linspace(min(x),max(x),10)
count = binning_1D(bins,x,5,2,1) 
binning = binning_1D(bins,x,5,2,y)

smooth = binning/count
plt.plot(bins,smooth,'r',lw=2)

c = np.column_stack((np.asarray(x),np.asarray(y)))
data = pd.DataFrame(c, columns=['Effectifs en moyens','Depassements moyens'])


sns.jointplot('Effectifs en moyens','Depassements moyens', data=data, kind="reg", color="r", size=10)
    
#plt.legend(loc='upper right',ncol=3)
```

Il n'y a pas assez de données pour les départements ayants des effectifs moyens > 60.
Pour les départements dont l'effectif moyen est < 60 : là où les effectifs sont les plus faibles, les dépassments ont tendance à être plus élevés.
