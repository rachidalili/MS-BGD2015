```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
def get_soup_for_page(url_base):
    page = requests.get(url_base)
    soup = BeautifulSoup(page.text, 'html.parser')
    return soup
```

```python
def get_products_for_page(num_page):
    url_base = 'http://www.cdiscount.com/search/10/acer.html?&page='+str(num_page)
    cdiscountSoup = get_soup_for_page(url_base)
    list_of_products = soup.find_all('div', { "class":"prdtBloc"})
    for product in list_of_products:
        link = product.find_all('a')[0].get('href')
        photo = product.find_all('a')[0].find_all('img')[0].get('data-src')
        name = product.find_all('a')[0].find_all('div', {"class":"prdtBTit"})[0].text
        price = product.find_all('span', {"class":"price"})[0].text
        oldPrice = price
        if len(product.find_all('div', {"class":"prdtPrSt"})):
            oldPrice = product.find_all('div', {"class":"prdtPrSt"})[0].text

        print name + " " + price + " (" + oldPrice + ")"
        print link
        print photo
        print "================================================="
    print "~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~"
        
```

```python
URL = 'http://www.cdiscount.com/search/10/acer.html'
soup = get_soup_for_page(URL)
page_numbers = soup.find_all('div', { "class":"pg"})[0].find_all('a')
max_page = page_numbers[len(page_numbers)-2]
print "There is more than " + max_page.text + " pages"


for idx in range(1, min(11, int(max_page.text) + 1 )):
    get_products_for_page(idx)
        
        
```
