```python
country_rex = re.compile(r'.*_(\w*)_([\w +]*)_.*[_]')
site_rex = re.compile(r'(.*)\|')
```

```python
def extract_country(name):
    return name.split('_')[1]

def extract_target(name):
    return name.split('_')[2]

def extract_site(name):
    try:
        return site_rex.findall(name)[0].strip().upper()
    except:
        return ''
```

```python
dat['Country'] = dat.campaign.apply(extract_country)
```

```python
dat['Target'] = dat.campaign.apply(extract_target)
```

```python
dat['Site'] = dat.article_title.apply(extract_site)
```

```python
dat.groupby(['Country', 'Target', 'Site'])['impressions', 'total_revenue'].mean()
```
